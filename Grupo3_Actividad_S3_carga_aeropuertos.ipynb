### Revisión de nulos

```python
for col in df_grupo3_aeroports.columns:
    print("Revisando nulos: ", col)
    print(df_grupo3_aeroports[col].isnull().sum())
```

### Revisión de nulos en col 'cod_iata_faa'

```python
df_grupo3_aeroports[ df_grupo3_aeroports['cod_iata_faa'].isnull()]
```

### Reemplazo nulos cod_iata_faa (3 pos - Z99)

```python
df_grupo3_aeroports['cod_iata_faa'] = df_grupo3_aeroports['cod_iata_faa'].fillna('Z99')
```

### Revisión de nulos en col 'cod_icao'

```python
df_grupo3_aeroports[ df_grupo3_aeroports['cod_icao'].isnull()]
```

### Reemplazo nulos cod_icao (4 pos - Z999)

```python
df_grupo3_aeroports['cod_icao'] = df_grupo3_aeroports['cod_icao'].fillna('Z999')
```

### Comprobaciones integridad país

### Para realizar la comprobación de integridad consultamos los datos de países en bd

```python
t_pais = %sql select cod_pais as cod_pais_p, pais from DM_PAIS
```

```python
df_grupo3_pais = t_pais.DataFrame()
df_grupo3_pais.head(5)
```

### Cruce aeropuerto - paises

```python
df_grupo3_aeroports.head(5)
```

```python
df_grupo3_aeroports_merged = pd.merge(left=df_grupo3_aeroports, right=df_grupo3_pais,
                      how='left', left_on='cod_pais', right_on='cod_pais_p')
```

```python
df_grupo3_aeroports_merged.head(10)
```

```python
df_grupo3_aeroports_merged[df_grupo3_aeroports_merged['cod_pais_p'].isnull()]
```

### Reemplazo nulos cod_pais_p (3 pos - Z99)

```python
df_grupo3_aeroports_merged['cod_pais_p'] = df_grupo3_aeroports_merged['cod_pais_p'].fillna('Z99')
```

### Añadimos una nueva columna: Concatenar 'cod_aeropuerto' +  'cod_pais'

```python
df_grupo3_aeroports_merged['JDL'] = df_grupo3_aeroports_merged['cod_aeropuerto'] + ' ' + df_grupo3_aeroports_merged['cod_pais_p']
```

```python
df_grupo3_aeroports_merged.head(5)
```

## Volcado del df a la bd

Nos quedamos con a estructura definitiva:     
COD_AEROPUERTO VARCHAR(4) NOT NULL,
NOMBRE_AEROPUERTO VARCHAR(100) NOT NULL,
CIUDAD_AEROPUERTO VARCHAR(45) NULL,
PAIS_AEROPUERTO VARCHAR(45) NULL,
COD_PAIS VARCHAR(3) NULL,
COD_IATA_FAA VARCHAR(3) NULL,
COD_ICAO VARCHAR(4) NULL,
LATITUD DECIMAL(14,8) NULL,
LONGITUD DECIMAL(14,8) NULL,
ALTITUD DECIMAL(14,8) NULL,
ZONA_HORARIA INT NULL,
DST VARCHAR(1) NULL,
ZONA_HORARIA_TZ VARCHAR(45) NULL

### Seleccion de columnas

```python
df_grupo3_aeroports_to_database = df_grupo3_aeroports_merged.drop(['cod_pais_p','pais', 'JDL'],1)
```

### Volcado a BBDD

```python
df_grupo3_aeroports_to_database.to_sql('DM_AEROPUERTOS', con = conn, if_exists = 'append', index=False)
```
